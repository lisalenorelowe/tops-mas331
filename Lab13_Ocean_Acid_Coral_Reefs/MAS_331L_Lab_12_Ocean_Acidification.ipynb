## Exercise 1: Understanding the marine CO2 system

The marine CO2 system is incredibly important and can have important implications:

 - Influences pH and controls the buttering capactiy of seawater
 - Carbon source for photosynthesis
 - Long-term storage of carbon
     - the ocean is a net sink for atmospheric CO2
 - Forms carbonate reefs
 - and much more!
 
 The CO2 system, or carbonate system, is the sum of dissolved inorganic carbon (DIC) species in seawater. **Total CO<sub>2</sub>** is equivalent to **total DIC** and is composed of **four main components** that are in independent equilibium with each other:
 
 - CO<sub>2</sub>
 - H<sub>2</sub>CO<sub>3</sub>
 - HCO<sub>3</sub><sup>-</sup>
 - CO<sub>3</sub><sup>2-</sup>
 
 The below equation describes the overall equilibrium of the carbonate system. Note that the source of CO<sub>2</sub> in the ocean is through exchange with the atmosphere and that H<sup>+</sup> ions are generated in the reaction. 
 
 
 <center> CO<sub>2,(g)</sub><=>CO<sub>2,(aq)</sub> + H<sub>2</sub>O <=> H<sub>2</sub>CO<sub>3</sub> <=> H<sup>+</sup> + HCO<sub>3</sub><sup>-</sup> <=> H<sup>+</sup> + CO<sub>3</sub><sup>2-</sup> </center>

One of the main controls of the marine carbonate system is the amount of CO<sub>2</sub> in the atmosphere. Before the Industrial Revolution, the concentration of CO<sub>2</sub> in the atmosphere was 280ppm; as of October 2022, that value 416ppm- a 49% increase! Because CO<sub>2</sub> is a reactant in carbonate equilibrium reaction, more CO<sub>2</sub> in the atmosphere drives the equilibrium of the reaction to the right, generating more products. The reverse would also be true- remove CO<sub>2,(g)</sub>, and the equilibrium is shifted to the left. 

*One of the key products are H<sup>+</sup> ions.* By now, we should all be familiar with the concept of pH. Remember, pH = -log([H<sup>+</sup>]), so it's an *inverted log scale.* The greater the amount of H<sup>+</sup> ions, the *lower* the pH.
So, the greater the amount of CO<sub>2</sub> dissolved in the ocean, the greater the amount of H<sup>+</sup> ions are generated, and lower the pH.

The total DIC varies depending on location and depth. In general, the Pacific Ocean is more enriched in DIC relative to the Atlantic Ocean. This is because the water in the Pacific is much older, and thus has had more time to absorb CO<sub>2</sub>. It also acculates from biological activites like respiration and the dissolution of CaCO<sub>3</sub>. Additionally, photosynthesis consumes CO<sub>2</sub> in the surface waters, so deeper water is more enriched as well. Take a look at Figure 1 for a visualization.

<center> Figure 1: Total DIC vs. depth for the Pacific and Atlantic Oceans. </center>

The total DIC in the ocean is also related to **alkalinity**. Simply put, alkalinity is a measure of the buffering capacity of seawater. It is the amount of *negative charge in seawater that is also able to accept a proton.* It is **not** the sum total of negative charge- there are negatively charged species, such as Cl<sup>-</sup>, that are unable to accept protons.

There are many different chemical species capable of accepting protons. We can *estimate* total alkalinity (TA) based on the chemical species HCO<sub>3</sub><sup>-</sup>, CO<sub>3</sub><sup>2-</sup>, B(OH)<sub>4</sub><sup>-</sup>, OH<sup>-</sup>, and H<sup>+</sup>. Because TA is based on the number of protons a species can accept, those such as CO<sub>3</sub><sup>2-</sup> that can accept multiple protons should be multiplied by the number they can accept. About 95% of the TA comes from the concentration of HCO<sub>3</sub><sup>-</sup> and CO<sub>3</sub><sup>2-</sup> and about 5% from B(OH)<sub>4</sub><sup>-</sup>. 


<center> TA = [HCO<sub>3</sub><sup>-</sup>] + 2[CO<sub>3</sub><sup>2-</sup>] + [B(OH)<sub>4</sub><sup>-</sup>] + [OH<sup>-</sup>] - [H<sup>+</sup>]
    

There are many other weak acids and bases that also influence the TA of seawater (e.g., HPO<sub>4</sub><sup>2-</sup>, SiO(OH)<sub>3</sub><sup>-</sup>, HS<sup>-</sup>, etc.). However, these are often in such low concentrations that we can ignore them and still get a good estimate of TA. The greater the alkalinity, the better the buffering ability.

Two of the most well-known examples are the Hawaiian Ocean Time-series (HOT) and the Bermuda Atlantic Time-series Study (BATS). The data from the time-series can be found online. ## Exercise 3: Plotting trend lines

We can see from the plots above that there are visually apparent trends for pH and pCO2. But can we quantify those trends? Let's use a simple linear regression to quantify the trends, as we discussed in lecture.


1. Fit the pH data from HOT to a linear regression using the code below. **Read the comments to understand what the code is doing**

Note that  we are using datenum (our serialized datetimes) for our x variable. Fitting a regression will not work with a datetime object. We need to arrange the x variable into an array with shape (-1, 1). We need to reshape the array because for the regression model to run, the x variable must have one column and as many rows as necessary. 

`x = np.array(HOT['datenum']).reshape((-1,1)) # define x as your datenum variable

Let's also run these lines of code, which will give us our R2 value, which is the coefficient of determination (how much variability the model explains), and our slope and intercept for our regression model line.

`r_sq = modelpH.score(x, y) # define r_sq as the score of your regression model, which is the coefficient of determination
print(f"coefficient of determination: {r_sq}") # print the R2 value of the regression model
print(f"intercept: {modelpH.intercept_}") # print the intercept of the regression model, which is the b term in y = mx+b
